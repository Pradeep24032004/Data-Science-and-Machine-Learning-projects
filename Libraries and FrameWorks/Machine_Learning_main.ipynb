```python
#data cleaning and preprocessing(done by using the pandas library)
from sklearn.datasets import fetch_openml
import pandas as pd
df = fetch_openml('titanic',version=1,as_frame=True)['data'] #converting data into dataframe
```

```python
df.info()
df.isnull().sum()
```

```python
#visualize using seaborn and visualize the null values
#calculating missing values percentage
#dividing the no of null values by total rows * 100
miss_val_per = pd.DataFrame((df.isnull().sum()/len(df))*100)
#visialize this miss_val_per
import seaborn as sns
sns.set()
miss_val_per.plot(kind='bar',title='missing values in percentage',ylabel='Percentage')
```

```python
#as body column has highest number of null values greater than 90% we are going to drop it
df.drop(['body'],axis=1)
```

```python
#instead of deleting the column we are going to replace the null values by using value imputation
#value Imputation:
'''
Simple Imputer is a very convinient strategy for missing data imputation it replaces all missing values with statistic calculated from the other values in the column use statistic are mean,median and mode '''

from sklearn.impute import SimpleImputer
print("Current null values", df['age'].isnull().sum())
```

```python
#method 1
imp = SimpleImputer(strategy='mean')
df['age'] = imp.fit_transform(df[['age']])
print("Present null values after imputation",df['age'].isnull().sum()) #all the null values are replaced with some values
```

```python
#method 2
def get_parameters(df):
    parameters = {}
    for column in df.columns[df.isnull().any()]:
        if df[column].dtype == 'float64' or df[column].dtype == 'int64' or df[column].dtype == 'int32':
            strategy = 'mean'
        else:
            strategy = 'most_frequent'  # mode
        missing_values = df[column][df[column].isnull()].values[0]
        parameters[column] = {'missing_values': missing_values, 'strategy': strategy}
    return parameters

# Example usage
parameters = get_parameters(df)
print(parameters)
```

```python
from sklearn.impute import SimpleImputer
parameters = get_parameters(df)
for column, param in parameters.items():
    missing_values = param['missing_values']
    strategy = param['strategy']
    imp = SimpleImputer(missing_values=missing_values, strategy=strategy)
df[column] = imp.fit_transform(df[[column]])
```
